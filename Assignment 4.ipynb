2. Display the Glucose level and the Age for the samples having the index ranging from 45 to 60

3. How many patients in the data are having the Blood Pressure level above 120 ?

```python
df=pd.read_csv('diabetes_model.csv')
bp=df[df['BloodPressure']>120]
c=bp.shape[0]
print("BP above 120:\t",c)
```

4. Find the minimum and maximum value of Insulin level in thedata

```python
df=pd.read_csv('diabetes_model.csv')
min=df['Insulin'].min()
max=df['Insulin'].max()
print("Minimum Insulin level: ",min)
print("\nMaximum Insulin level: ",max)
```

5. Create a histogram of Age

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('diabetes_model.csv')
plt.hist(df['Age'])
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.show()
```

6. Create a scatter plot of BMI vs Glucose

```python
df=pd.read_csv('diabetes_model.csv')
plt.scatter(df['BMI'],df['Glucose'],color='red')
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.show()
```
